# コンピューターの能力

## プログラミングとは？
"Program"...Pro(先に) + gram(書く)  
コンピューターに対する先出しの**命令**

四則演算

```python
#print(x)は「xを表示」
print(33+4)
print(33-4)
print(33*4)
print(33/4)
```

値の記憶

```python
x = 33 #xに33を記憶
x *= 4  #xに4を掛け算
x += 226 #xに226を足し算
print(x) #xの中身を表示
```

条件分岐

```python
x = 5 #xに5を記憶

if x > 3:
    print("xは3より大きい")
else:
    print("xは3以下")
    
if x > 10:
    print("xは10より大きい")
else:
    print("xは10以下")
```

繰り返し

```python
x = 15  #xに15を記憶

#99回繰り返し
for cnt in range(99):
    x += 4 #xに4を足し算
    
print(x) #xの中身を表示
```

## 計算の速さ

```python
from time import time

loop_times = 10000000 #繰り返し回数

time_start = time() #(時間計測用)

#繰り返し
for cnt in range(loop_times):
    x = 25 * 25 * 3.14 #半径25cmの円の面積を求める
    
time_end = time() #(時間計測用)

#計測結果を表示
time_delta = time_end - time_start
print(f"{time_delta}秒かかった")
per_one = time_delta / loop_times
print(f"1回あたり{per_one}秒かかった")
```

## イーロンマスクの最頻単語を求める
(C) Elon Musk

```python
import pandas as pd
```

```python
df = pd.read_csv("elonmusk_tweets.csv")

df.tail(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

time_start = time() #(時間計測用)

#数える
cv = CountVectorizer()
df_count = pd.DataFrame(cv.fit_transform(df["text"]).toarray(), columns=cv.get_feature_names_out())
sr_all = df_count.sum(axis=0)
vocabulary = sr_all.sort_values(ascending=False)

#計測結果を表示
time_end = time()
time_delta = time_end - time_start
print(f"{time_delta}秒かかった\n")

#数えた結果を表示
pd.set_option('display.max_rows', None)
print(vocabulary[0:1000])
```
